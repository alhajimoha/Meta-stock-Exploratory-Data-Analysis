```python
#let's check for Outliersusing boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(df['Close'])
plt.title('Box plot of closing price')
plt.ylabel('Closing Price')
plt.show()
```

```python
plt.figure(figsize = (10, 6))
plt.plot(df['Date'], df['Close'], color = 'blue')
plt.title("Closing price of Meta over time")
plt.xlabel("Date")
plt.ylabel("Closing price")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Close'], bins = 20, color = 'blue', edgecolor = 'black')
plt.title("Histogram of closing price")
plt.xlabel('Closing price')
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
#DAILY RETURNS
df["Daily returns"] = df["Close"].pct_change() * 100
print("Daily returns:")
print(df["Daily returns"].head())
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df["Daily returns"].dropna(), bins = 20, color='blue', edgecolor="black")
plt.title("Histogram of Daily returns")
plt.xlabel("Daily returns")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
window = 30
df["Rolling Mean"] = df["Close"].rolling(window=window).mean()
plt.figure(figsize=(10, 6))
plt.plot(df["Date"], df["Close"], label="Original", color="Blue")
plt.plot(df['Date'], df['Rolling Mean'], label= f'Rolling Mean({window} days)', color='red')
plt.title("Closing price and Rolling mean")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```
